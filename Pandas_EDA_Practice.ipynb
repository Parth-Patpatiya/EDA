```python
df=pd.read_csv('virat.csv')
df
```

```python
df['Runs']=pd.to_numeric(df['Runs'], errors='coerce')
df['Mins']=pd.to_numeric(df['Mins'], errors='coerce')
df['SR']=pd.to_numeric(df['SR'], errors='coerce')
```

```python
df.info()
```

```python
sns.displot(df['4s'], bins=10, color='r')
plt.show()
```

```python
plt.figure(figsize=(20,15))
```

```python
plt.figure(figsize=(20,20))
sns.displot(df['Dismissal'], bins=70, color='r')
plt.show()
```

```python
df['New_Column'] = np.linspace(0, 131, num=len(df))
df
```

```python
sns.jointplot(x=df['New_Column'], y=df['Runs'])
plt.show()
```

```python
plt.figure(figsize=(20,20))
df['Runs'].value_counts().plot.barh()
plt.show()
```

```python
plt.figure(figsize=(20,20))
df['SR'].value_counts().plot.bar()
plt.show()
```

```python
plt.figure(figsize=(20,20))
df['Ground'].value_counts().plot.bar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Matches Played by Virat against the countries')
plt.xlabel('Number of Matches')
plt.ylabel('Oppositions')
plt.legend()
df['Opposition'].value_counts().plot.barh(edgecolor='orange', color='cyan')
plt.show()
```

### In which yeart Kohli scored max runs

```python
plt.bar(df['year'],df['Runs'],width=0.5,align='center',edgecolor='red',color='green')
plt.show()
```

```python
plt.figure(figsize=(50,50))
sns.pairplot(data=df, x_vars=['year'],y_vars='Runs')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='year', y='Runs', data=df,  color='purple')
plt.title("Virat Kohli's Yearly Runs")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.xticks(rotation=45)
plt.show()
```

```python
df_z=df.groupby(by='year')['Runs'].max()
plt.bar(df_z.index, df_z.values, width=0.5, align='center', edgecolor='red', color='green')

plt.xlabel('Year')
plt.ylabel('Max Runs')

plt.show()
```
